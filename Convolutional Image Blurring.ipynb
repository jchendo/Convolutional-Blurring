Convolutional Image Blurring

```python
import skimage as sk
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = sk.io.imread('C:/Users/jacob/Downloads/kiera.jpg')
plt.imshow(img)

sigma = 5
dimension_length = 9

img_red = img[:, :, 0]
img_green = img[:, :, 1]
img_blue = img[:, :, 2]

## A gaussian kernel is just a weight matrix that gives values smaller weights as they increase in distance from the center.
kernel = cv2.getGaussianKernel(dimension_length,sigma) 
blurring_matrix =  kernel @ kernel.T ## multiply kernel (1D) by its transposition (1D) to create a 2D matrix, a true blurring matrix
blurred_img_red = signal.convolve2d(img_red, blurring_matrix, mode='same')
blurred_img_green = signal.convolve2d(img_green, blurring_matrix, mode='same')
blurred_img_blue = signal.convolve2d(img_blue, blurring_matrix, mode='same')

## Stacking to combine values w/o reordering.
blurred_img_2D = np.stack((blurred_img_red, blurred_img_green, blurred_img_blue), axis=-1).astype(int)
plt.imshow(blurred_img_2D)
```
