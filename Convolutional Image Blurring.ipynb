Convolutional Image Blurring

```python
import skimage as sk
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

def surrounding_pixels(matrix, row, col):
    pixels = np.empty(9, dtype=int)
    for i in range(-1, 2):
        for j in range(-1, 2):
            r = int(row + i)
            c = int(col + j)
            try: ## out of bounds 
                print(matrix[r][c])
                pixels[r+c] = matrix[r][c]
                print(pixels[r+c])
            except:
                continue
    return pixels
```

```python
kiera = sk.io.imread('C:/Users/jacob/Downloads/kiera.jpg')
plt.imshow(kiera)
#kiera = kiera.reshape(249795, 3)
for convolution_num in range(1): # num pixels, generalize later
    row_num = np.floor((convolution_num+1)/585)
    col_num = (convolution_num - (row_num * 585))
    pixels = surrounding_pixels(kiera, row_num, col_num)
    print(pixels)

blurring_matrix = np.array([[1/9, 1/9, 1/9], [1/9, 1/9, 1/9], [1/9, 1/9, 1/9]])
blurred_img_2D = signal.convolve2d(kiera, blurring_matrix, mode='same')
blurred_img_2D = blurred_img_2D.reshape(427, 585, 3).astype(int)
plt.imshow(blurred_img_2D)
```
